# Filtrando y agrupando datos

## Metodo simple

Podemos filtrar la base de datos en base a condiciones sobre una o mas de sus columnas. Para esto usamos la siguiente sintaxis: 

```data[condicion que queremos que se cumpla]```

Las condiciones se veran de la siguiente forma

* ```data[columna] == valor```
* ```data[columna] > valor```
* ```data[columna] =! valor```
* ```data[columna].isin(lista de valores)```

````python
# Vinos de Estados Unidos

data[data["country"] == 'US']
````

````python
# Vinos que valen mas de 50 dolares

data[data["price"] > 50]
````

````python
# Vinos de Chile y Argentina

data[data["country"].isin(["Chile", "Argentina"])]
````

## Metodo query de pandas

Otra manera de filtrar los datos es usando el metodo query de pandas 

* Documentacion: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html
* Buen post con ejemplos: https://towardsdatascience.com/10-examples-that-will-make-you-use-pandas-query-function-more-often-a8fb3e9361cb

En este caso las condiciones se escriben como un gran string y se pueden mezclar facilmente entre ellas

````python
data.query('country == "Chile" and points > 85')
````

## Agrupando datos

Para sacar metricas por datos agrupados podemos usar el group by. Podemos aplicar operaciones de: 

* suma
* media
* conteo

````python
data.groupby(['country'])["price"].mean()
````

````python
data.groupby(['country']).count()
````
