# Filtrando y agrupando datos

## Filtrar: metodo simple

Podemos filtrar la base de datos en base a condiciones sobre una o mas de sus columnas. Para esto usamos la siguiente sintaxis: 

```data[condicion que queremos que se cumpla]```

Las condiciones se veran de la siguiente forma

* ```data[columna] == valor```
* ```data[columna] > valor```
* ```data[columna] =! valor```
* ```data[columna].isin(lista de valores)```

````python
data["country"].unique()
````

````python
# Vinos de Estados Unidos

# asignamos la tabla filtrada al nombre vinos_us
vinos_us = data[data["country"] == 'US']

# vemos los valores distintos de la columna pais
print(vinos_us["country"].unique())
````

````python
vinos_us.describe(include="all")
````

````python
# Vinos que valen mas de 50 dolares

vinos_50mas = data[data["price"] > 50]
````

````python
vinos_50mas.describe()
````

````python
# Vinos de Chile y Argentina

vinos_CAI = data[data["country"].isin(["Chile", "Argentina", "Italy"])]
vinos_CAI["country"].unique()
````

## Metodo query de pandas

Otra manera de filtrar los datos es usando el metodo query de pandas 

* Documentacion: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html
* Buen post con ejemplos: https://towardsdatascience.com/10-examples-that-will-make-you-use-pandas-query-function-more-often-a8fb3e9361cb

En este caso las condiciones se escriben como un gran string y se pueden mezclar facilmente entre ellas

````python
vinos_buenos_chile = data.query('country == "Chile" and points > 85')
````

````python
vinos_buenos_chile.describe(include="all")
````

## Agrupando datos

Para sacar metricas por datos agrupados podemos usar el group by. Podemos aplicar operaciones de: 

* suma
* media
* conteo

````python
# Calcular el precio promedio de vinos por pais

data.groupby(['country'])["price"].mean()
````

Podemos hacer lo mismo que hace el group by con otros metodos, pero usar group by nos facilita la vida

````python
# Alternativa 1: hacer esto para cada pais

data[data["country"]== "Chile"]["price"].mean()
````

````python
# Alternativa 2: hacer lo de arriba con un for

paises = list(data["country"].unique())
````

````python
for pais in paises:
    print(pais, data[data["country"]==pais]["price"].mean())
````
