## Clasificacion

Para el caso de clasificacion ...

```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,15)
```

```python
iris = load_iris()
clf = tree.DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(iris.data, iris.target)
```

```python
_ = tree.plot_tree(clf)
```

## Regresion

En el caso de regresion ... 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True)
regressor = DecisionTreeRegressor(random_state=0, max_depth=3)

regressor.fit(X,y)
```

```python
_ = tree.plot_tree(regressor)
```

## Modelos asociados: Random Forest

Existen algunos modelos asociados o basados en arboles, el mas comun de ellos es el random forest que consiste en un conjunto de arboles de decision, cada uno entrenado con una porcion de los datos y/o atributos del dataset original. Para hacer sus predicciones, el random forest considera las predicciones de cada uno de los arboles que lo componen y

* en el caso de clasificacion, produce como output la clase mas votada
* en el caso de regresion, produce como output el promedio de todas las predicciones

Hiperparametros mas relevantes: 
* n_estimators: 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X,y)
```
